```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
```

```python
#Load the data
train_data = pd.read_csv('Color_Classification.csv')
train_data
```

```python
def convert_int(vector):
    int_vector = np.ones(len(vector))
    values = set(vector)
    cat_int = 0
    for val in values:
        int_vector[vector==val] = cat_int
        cat_int += 1
    return int_vector
```

```python
train_data['cat_int'] = convert_int(train_data['Y'])
train_data.head()
```

```python
#Plot
plt.figure()
plt.scatter(train_data['X'],train_data['cat_int'])
plt.xlabel('x value')
plt.show()
```

```python
#Fit the data with Linear Regression Model
LR = LinearRegression(fit_intercept=True)
LR.fit(train_data[['X']],train_data['cat_int'])
predictions = LR.predict(train_data[['X']])
train_data['pred']=predictions.round()
accuracy_score(train_data['cat_int'],train_data['pred'])
```

```python
#Plot the predictions and the sample outputs y 作图对比pred和Y的真实值
plt.scatter(train_data['X'],train_data['pred'], color='b', linestyle='dotted', label='Prediction')
plt.scatter(train_data['X'],train_data['cat_int'], color='r', linestyle='dotted', label='Category')
plt.xlabel("data index")
plt.title("Comparison between model predictions and the true Y.")
plt.legend(loc='upper right')
plt.show()
```

```python
#############However, if we change the labels, the predictions become bad
#Blue=1, Orange=2, Red=3
int_vector = np.ones(len(train_data))
int_vector[train_data['Y']=='Blue'] = 1
int_vector[train_data['Y']=='Orange'] = 2
int_vector[train_data['Y']=='Red'] = 3
train_data['cat_int'] = int_vector

#Plot
plt.figure()
plt.scatter(train_data['X'],train_data['cat_int'])
plt.xlabel('x value',color='white')
plt.ylabel('y value',color='white')
plt.tick_params(colors='white')
plt.show()
```

```python
#Fit the data with Linear Regression Model
LR = LinearRegression(fit_intercept=True)
LR.fit(train_data[['X']],train_data['cat_int'])
predictions = LR.predict(train_data[['X']])
train_data['pred']=predictions.round()
accuracy_score(train_data['cat_int'],train_data['pred'])
```

```python
#Plot the predictions and the sample outputs y 作图对比pred和Y的真实值
plt.scatter(train_data['X'],predictions, color='b',  label='Fitted Line')
plt.scatter(train_data['X'],train_data['cat_int'], color='r', label='Data')
plt.xlabel("data index")
plt.title("Data and fitted line.",color='white')
plt.xlabel('x value',color='white')
plt.ylabel('y value',color='white')
plt.tick_params(colors='white')
plt.legend(loc='upper left')
plt.show()
```
